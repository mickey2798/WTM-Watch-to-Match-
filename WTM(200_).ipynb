```python
# Import Pandas
import pandas as pd

# Load teacher Metadata
metadata = pd.read_csv('C:\\Users\\Piyush_Kumar_Sahoo\\Desktop\\wtm.csv')

# Print the first three rows
metadata.head(5)
```

```python
sp=input()   #It takes the specialisation as an input from the user(Content-Based filtering)
```

```python
df2=metadata.copy().loc[metadata['specialisation'] == sp]   #Preparing a dataframe related to your desired Specialization
```

```python
df2 #It consists of all those teachers having the user defined specialisation
```

```python
#Normalizing all the essential Factors or the determinants and then scaling them in a range between 0-10
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
column_names_to_normalize = ['Years of experience', 'R1(CS)', 'R2(Knowledge)', 'R3(Interest)', 'R4(Way)','R5(Overall)','Project','Research', 'Certification']
x = df2[column_names_to_normalize].values
x_scaled = min_max_scaler.fit_transform(x)
df_temp = pd.DataFrame(x_scaled, columns=column_names_to_normalize, index = df2.index)
df2[column_names_to_normalize] = df_temp*10
```

```python
df2
```

```python
#Number of characteristics you want to take up.
print('How many characteristics you wanna compare??')
n = int(input())
```

```python
#Provide all your desired choices.
a=[]
print("Enter your choices")
for i in range(n):
    i = input()
    a.append(i)
```

```python
# creating a dataframe containing the teachers along with their data regarding the user defined characteristics 
df = df2[a]
df.insert(0, "Teacher", df2['Teacher'], True) #inserting teacher data in the new dataframe.
df.insert(0, "T_id", df2['t_id'], True)
df
```

```python
#Calculating the score for each teacher on the basis of the user defined characteristics
df['score'] = df[a].mean(axis=1)
```

```python
df
```

```python
#Applying KNN
df['New'] = 10
df['New'] = pd.to_numeric(df['New'], errors='coerce')
```

```python
df
```

```python
#Finding the Manhattan distance
df['distance'] = abs(df['score']-df['New'])
```

```python
df
```

```python
#Sorting the dataframe on the basis of distance
Teacher_Recommendation = df.sort_values(by=['distance'])
```

```python
#Number of teachers Required
k=int(input())
```

```python
Result= Teacher_Recommendation.head(k)
```
